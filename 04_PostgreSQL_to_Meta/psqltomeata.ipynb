```python
import pandas as pd
from sqlalchemy import create_engine, inspect, text
```

```python
#pip install psycopg2-binary # PostgreSQL veritabanı bağlantısı için gerekli kütüphane
```

```python
DB_CONNECTION = 'postgresql://user:password@localhost:5432/metabase_app'
```

```python
engine = create_engine(DB_CONNECTION)
inspector = inspect(engine)
table_names = inspector.get_table_names()
```

```python
print(table_names)
```

```python
df=pd.read_csv("Sonhali.csv")
```

```python
with engine.connect() as conn:
    for table in table_names:
        query = text(f'SELECT COUNT(*) FROM "{table}"')
        result = conn.execute(query)
        count = result.scalar()
        print(f"{table:<30} | {count}")
```

```python
df.to_sql('Satislar', engine, if_exists='replace', index="Tweet_kimligi", chunksize=1000)
```
